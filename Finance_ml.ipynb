```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('http://hilpisch.com/pyalgo_eikon_eod_data.csv',index_col=0,parse_dates=True).dropna()
```

```python
df
```

```python
x=df["EUR="][:2500].to_numpy()
data=x.reshape(-1,10)
    
```

```python
data.shape
```

```python
x=data[:,:1]
y=data[:,-1:]
```

```python
y
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x[:200],y[:200])
```

```python
predict=lr.predict(x[200:])
predict.shape
```

```python

plt.plot(np.ndarray.flatten(y[200:]))
plt.plot(np.ndarray.flatten(predict),"bx")
```
